```python
df=pd.read_csv("SalaryData_Test(1).csv")
```

```python
label_encoder = preprocessing.LabelEncoder()
df['workclass']= label_encoder.fit_transform(df['workclass'])
df['occupation']= label_encoder.fit_transform(df['occupation'])
df['Salary']= label_encoder.fit_transform(df['Salary'])
```

```python
df=df.drop(['education','educationno','maritalstatus','relationship','race','sex','capitalgain','capitalloss','native'],axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.iloc[:,0:4]
Y=df['Salary']
```

```python
X
```

```python
Y
```

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
model.y_pred=model.predict(X_test)
```

```python
model.y_pred
```

```python
rmse=np.mean((model.y_pred-Y_test)**2)
```

```python
print("rmse on test set : ",rmse)
```
